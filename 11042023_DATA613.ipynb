```python
# 1. (1 pt) Use relative paths to load these data frames into Python.

import pandas as pd
import numpy as np


country = pd.read_csv("../DATA613/hw-09-python-Ryo-TANAKA-programmer/dt/country.csv")
fer = pd.read_csv("../DATA613/hw-09-python-Ryo-TANAKA-programmer/dt/fertility.csv")
life_exp = pd.read_csv("../DATA613/hw-09-python-Ryo-TANAKA-programmer/dt/life_exp.csv")
pop = pd.read_csv("../DATA613/hw-09-python-Ryo-TANAKA-programmer/dt/population.csv")
```

```python
new_pop = pd.melt(pop, 
                  id_vars = ["Country Name", "Country Code"], 
                  var_name = "year", 
                  value_name = "pop")
new_fer = pd.melt(fer, 
                  id_vars = ["Country Name", "Country Code"],
                  var_name = "year", 
                  value_name = "fer_rate")
```

```python
new_life_exp = pd.melt(life_exp, 
                       id_vars = ["Country Name", "Country Code"],
                       var_name = "year", 
                       value_name = "life_exp")
```
