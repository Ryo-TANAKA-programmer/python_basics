```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
# sns.kdeplot(data you want to pass)
# ;(semicolon) means seperation
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower,shade = True)
```

```python
sns.kdeplot(cars.horsepower,bw_method = 1);
```

```python
sns.kdeplot(cars.horsepower,cumulative = False);
```

```python
sns.set()
sns.kdeplot(cars.horsepower,cars.mpg);
```

```python
sns.kdeplot(cars.horsepower,cars.mpg,n_levels = 10,
           shade_lowest = False,
           cbar = True);
```
