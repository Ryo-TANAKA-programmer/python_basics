```python
# standard imports
import numpy as np
import pandas as pd
from numpy.random import randn
```

```python
# stats import
from scipy import stats
```

```python
# plotting import
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# %matplotlib inline allows the jupyter notebook to display
%matplotlib inline 
```

```python
# normal distribution
dataset1 = randn(500)
```

```python
# plt.hist makes a histogram
sns.set()
plt.hist(dataset1) 
```

```python
dataset2 = randn(400)

plt.hist(dataset2,color = 'red')
```

```python
# density = True is important since dataset1 and dataset2 have different totals.
# bins are the number of bins obviously

plt.hist(dataset1,density = True,color = 'blue',
        alpha = 0.15,bins = 25)
plt.hist(dataset2,density = True,color = 'red',
        alpha = 0.15,bins = 25)
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sns.jointplot(data1,data2, kind = 'scatter')
plt.savefig('jointplot.png')
```
