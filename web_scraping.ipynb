```python
import requests
baseurl_brewery = "https://api.openbrewerydb.org/v1/breweries"
```

```python
params = {
    "by_state": "Ohio",
    "by_type": "micro",
    "per_page": 200
}

# making a GET request
brewery_out = requests.get(baseurl_brewery, params = params)
```

```python
import pandas as pd
import json

# Parse the JSON response using brewery_out
brewery_data = brewery_out.json()

# making brewery_df using pd.DataFrame 
brewery_df = pd.DataFrame(brewery_data)

brewery_df.head()
```

```python
brewery_df_filtered = brewery_df[~(brewery_df["longitude"].isna() | brewery_df["latitude"].isna())]
brewery_df_filtered.tail()
```

```python
print(brewery_df_filtered["name"])
```

```python
# now using another api

eng_women_html_url = "https://data-science-master.github.io/lectures/data/engart.html"
response = requests.get(eng_women_html_url)
print(response)
```

```python
if response.status_code == 200:
    html_content = response.text
else:
    print("error")
```

```python
from bs4 import BeautifulSoup
```

```python
eng_soup = BeautifulSoup(html_content, "html.parser")
```

```python
selectors = [
    "ul:nth-child(49) li , ul:nth-child(47) li, ul:nth-child(45) li, ul:nth-child(43) li, ul:nth-child(41) li, ul:nth-child(39) li, ul:nth-child(37) li, ul:nth-child(35) li, ul:nth-child(33) li, ul:nth-child(31) li, ul:nth-child(29) li, ul:nth-child(27) li, ul:nth-child(25) li, ul:nth-child(23) li, ul:nth-child(21) li, ul:nth-child(19) li, ul:nth-child(17) li, ul:nth-child(15) li, ul:nth-child(13) li, ul:nth-child(11) li, ul:nth-child(9) li, ul:nth-child(7) li, ul:nth-child(5) li"
]
eng_women = []
```

```python
for selector in selectors:
    elements = eng_soup.select(selector)
    for element in elements:
        eng_women.append(element.get_text())
```

```python
eng_women_df = pd.DataFrame(eng_women)
```

```python
eng_women_df.drop(columns = ["0"], axis = 1, inplace = True)
```

```python
eng_women_df.rename(columns = {0:"name"},
                    inplace = True)
```

```python
eng_women_df.columns = ["name"]
```

```python
eng_women_df.set_axis(["name"],
                       axis = 1,
                       inplace = True)
```

```python
eng_women_df
```

```python
eng_women_df[['new_name', 'other_info']] = eng_women_df["name"].str.split('(', n = 1, expand = True)
```

```python
eng_women_df[["year", "job"]] = eng_women_df["other_info"].str.split(")", 
                                                                     n = 1, 
                                                                     expand = True)
```
