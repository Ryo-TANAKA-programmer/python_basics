```python
data = {"rn":[1,1,2,2],
        "foo": ["A","B","A","B"],
        "bar": [5,6,7,8]}
df = pd.DataFrame(data)
print(df)
```

```python
df2 = df.pivot_table(index='rn', columns = 'foo', values='bar').reset_index()
```

```python
df2
```

```python
df
```

```python
df3 = df2.melt(id_vars = "rn",
         value_vars = ["A","B"],
         var_name = "foo",
         value_name = "bar").sort_values(by = "rn").reset_index(drop = True)
```

```python
df3
```

```python
data_1 = {"x": ["A","B","C","D"],
        "y": [1,2,3,4]}
data_1 = pd.DataFrame(data_1)
print(data_1)
```

```python
data_2 = {"z": ["A","C","E"],
          "w": [5,6,7]}
data_2 = pd.DataFrame(data_2)
print(data_2)
```

```python
pd.merge(left = data_1, 
         right = data_2,
         how = "left", left_on = "x", right_on = "z")
```

```python
pd.merge(left = data_1,
         right = data_2,
         how = "right",
         left_on = "x",
         right_on = "z")
```

```python
pd.merge(left = data_1,
         right = data_2,
         how = "inner",
         left_on = "x",
         right_on = "z")
```

```python
pd.merge(left = data_1,
         right = data_2,
         how = "outer",
         left_on = "x",
         right_on = "z")
```
